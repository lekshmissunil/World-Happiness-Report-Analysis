```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

# World Happiness Report Analysis

```python
data = pd.read_csv('../input/world-happiness-report-2021/world-happiness-report-2021.csv') 
data.head()
```

# **Overall Top 5 & Bottom 5 Countries**

```python
t=data.head() 
b=data.tail()  
d = pd.concat([t,b], axis=0)
sns.barplot(data = d, x = 'Country name', y = 'Ladder score')
plt.xticks(rotation=90) 
plt.title("World Happiness Report: Top & Bottom 5 countries") 
plt.show()  
```

# **Correlation between criteria**

```python
fig_dims = (20,8)
fig, ax = plt.subplots(figsize = fig_dims)
sns.heatmap(data[['Ladder score', 'Logged GDP per capita','Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption'] ].corr(), annot = True) 
plt.show()
```

# **Regional Indicators**

```python
d = data.groupby("Regional indicator").mean()['Logged GDP per capita'].reset_index() 
d= d.sort_values('Logged GDP per capita',ascending=False)
sns.barplot(data = d, x = 'Regional indicator', y = 'Logged GDP per capita', color='orange')
plt.xticks(rotation=90) 
plt.title("Logged GDP per capita by Region") 
plt.show()
```

```python
d = data.groupby("Regional indicator").mean()['Social support'].reset_index() 
d= d.sort_values('Social support',ascending=False)
sns.barplot(data = d, x = 'Regional indicator', y = 'Social support', color='orange')
plt.xticks(rotation=90) 
plt.title("Social support by Region")
plt.show()
```

```python
d = data.groupby("Regional indicator").mean()['Healthy life expectancy'].reset_index() 
d= d.sort_values('Healthy life expectancy',ascending=False)
sns.barplot(data = d, x = 'Regional indicator', y = 'Healthy life expectancy', color='orange')
plt.xticks(rotation=90) 
plt.title("Healthy life expectancy by Region")
plt.show()
```

```python
d = data.groupby("Regional indicator").mean()['Freedom to make life choices'].reset_index() 
d= d.sort_values('Freedom to make life choices',ascending=False)
sns.barplot(data = d, x = 'Regional indicator', y = 'Freedom to make life choices', color='orange')
plt.xticks(rotation=90) 
plt.title("Freedom to make life choices by Region")
plt.show()
```

```python
d = data.groupby("Regional indicator").mean()['Perceptions of corruption'].reset_index() 
d= d.sort_values('Perceptions of corruption',ascending=False)
sns.barplot(data = d, x = 'Regional indicator', y = 'Perceptions of corruption', color='orange')
plt.xticks(rotation=90) 
plt.title('Perceptions of corruption by Region')
plt.show()
```

```python
d = data.groupby("Regional indicator").mean()['Generosity'].reset_index() 
d= d.sort_values('Generosity',ascending=False)
sns.barplot(data = d, x = 'Regional indicator', y = 'Generosity', color='orange')
plt.xticks(rotation=90) 
plt.title("Generosity by Region")
plt.show()
```
